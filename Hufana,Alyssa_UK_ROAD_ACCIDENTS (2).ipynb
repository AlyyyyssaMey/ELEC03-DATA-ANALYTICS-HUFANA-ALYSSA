```python
accidents_per_year = london_df['Year'].value_counts().sort_index()
accidents_per_month = london_df['Month'].value_counts().sort_index()
accidents_per_day = london_df['DayOfWeek'].value_counts().sort_index()


#PER YEAR 
accidents_per_year.plot(kind='bar')
plt.title('Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.show()

month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Replace numerical months with their names
accidents_per_month.index = accidents_per_month.index.map(month_names)


accidents_per_month.plot(kind='bar')
plt.title('Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()


#accident per day
day_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

accidents_per_day.index = accidents_per_day.index.map(day_names)
accidents_per_day.plot(kind='bar')
print(accidents_per_day)
```

```python
#AccidentsSeverityInLondon
severity_london  = london_df['Accident_Severity'].value_counts()

plt.pie(severity_london,
        labels = severity_london.index,
        autopct = '%1.1f%%',
        startangle = 90)
plt.gcf().set_size_inches(10,7)
plt.title('Accident Severity Distribution')
plt.show()

severity_2019_london = london_df[london_df['Year'] == 2019]
# severity_2019
lsc_2019 = severity_2019_london['Accident_Severity'].value_counts()

plt.pie(lsc_2019,
        labels = lsc_2019.index,
        autopct = '%1.1f%%',
        startangle = 90)
plt.gcf().set_size_inches(10,7)
plt.title('Accident Severity Distribution')
plt.show()
```

<h3> 26. What are the monthly accident trends in Birmingham, and how do they compare with national statistics? </h3>

```python
district = "Birmingham"
district_data = accidents[accidents['District Area'] == district].copy()
district_data['Accident Date'] = pd.to_datetime(district_data['Accident Date'])
district_data['Month'] = district_data['Accident Date'].dt.month
district_data['Year'] = district_data['Accident Date'].dt.year

# Pivot for heatmap
pivot = district_data.pivot_table(index='Month', columns='Year', values='Index', aggfunc='count')

plt.figure(figsize=(8,5))
sns.heatmap(pivot, cmap="Reds", annot=True, fmt="d")
plt.title(f"Monthly Accidents Heatmap - {district}")
plt.show()

# Compare with national monthly distribution
national_monthly = accidents.copy()
national_monthly['Accident Date'] = pd.to_datetime(national_monthly['Accident Date'])
nat_monthly = national_monthly.groupby(national_monthly['Accident Date'].dt.month)['Index'].count()
dist_monthly = district_data.groupby('Month')['Index'].count()

(dist_monthly / dist_monthly.sum()).plot(label=district)
(nat_monthly / nat_monthly.sum()).plot(label="National", linestyle="--")
plt.legend()
plt.title(f"{district} vs National Accident Distribution")
plt.show()
```

<h3> Insight #26 - Birmingham has the highest accident count nationally. It follows the national seasonal pattern but shows sharper winter peaks, likely due to traffic density and weather. </h3>
